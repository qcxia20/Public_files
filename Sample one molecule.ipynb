```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
# Generate one conformer
m1 = Chem.MolFromSmiles('N#Cc1ccccc1-c1cc(-c2ccccn2)cn(-c2ccccc2)c1=O')
m1
```

```python
# Generate one conformer
Chem.AddHs(m1) # Add H atoms
#Chem.RemoveHs(molecule_name) # Remove H atoms
m1
```

```python
# Generate one conformer
AllChem.EmbedMolecule(m1, params=AllChem.ETKDG()) # Embed a molecule with 3D coordinates; Use ETKDG-based distance geometry to generate one conformation
m1 # type rdkit.Chem.rdchem.Mol
```

```python
# Generate one conformer
AllChem.MMFFOptimizeMolecule(m1) # (not necessary with ETKDG) use MMFF94 force field to minimize an RDKit-generated cnformer
# At present, ETKDG is the default conformation generation method.
m1
```

```python
# Generate multiple(10 here) conformers
# run ETKDG 10 times to generate 10 conformers from 10 different random start points
# Set numThreads to 0 -> software to use the maximum number of CPU threads
m1 = Chem.MolFromSmiles('N#Cc1ccccc1-c1cc(-c2ccccn2)cn(-c2ccccc2)c1=O')
Chem.AddHs(m1) # type rdkit.Chem.rdchem.Mol
AllChem.EmbedMultipleConfs(m1, numConfs=10, params=AllChem.ETKDG())
for i in m1.GetConformers(): 
    print(i.GetPositions()) # print all 10 3D coordinations of one molecule in a list
    # 27 atoms for each sublist, corresponding to the 27 heavy atoms in total
```

```python
m1_4= m1.GetConformer(id=4) # type rdkit.Chem.rdchem.Conformer
print(m1_4.GetPositions()) # print one of the 3D coordinate from one conformation
print(m1_4.Is3D()) # print the Bool Value of 
```
